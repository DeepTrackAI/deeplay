```python
from typing import Dict, Any, Union, Tuple, overload

from deeplay import DeeplayModule, Layer
import torch.nn as nn

class FromDict(DeeplayModule):
    def __init__(self, t, *keys: str, **kwargs):
        super().__init__()
        self.keys = keys

    def forward(self, x: Dict[str, Any]) -> Union[Any, Tuple[Any, ...]]:
        return (

            x[self.keys[0]]
            if len(self.keys
                   ) == 1
            else tuple(x[key] for key in self.keys)
        )
```

```python
module = FromDict(3, "a", "b", arg=3)
built = module.build()

print(built({"a": 1, "b": 2}))
```
